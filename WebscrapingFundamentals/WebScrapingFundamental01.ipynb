### get the douban.com as practice

```python
url='https://movie.douban.com/review/best/'
#use the avoid scarping mechanism to scrape
```

```python
import requests
def getHTMLtext(url):
    try:
        r=requests.get(url,timeout=20)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return'error'
if __name__=='__main__':
    url='https://movie.douban.com/review/best/'
    print(getHTMLtext(url))
```

```python
import requests
from lxml import etree
r=requests.get(url).text
s=etree.HTML(r)
a=s.xpath("//div[@class='short-content']/text()")
print(a)
```
